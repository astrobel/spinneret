```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
t = np.linspace(0, 1, 1000, False)  # 1 second
sig = np.sin(2*np.pi*10*t) + np.sin(2*np.pi*20*t)

%matplotlib notebook
fig, ax1 = plt.subplots(1, 1, sharex=True)

ax1.plot(t, sig, 'r', label='true underlying signal', lw=3)

sig += 10*t + 50*(t-0.5)**2
ax1.plot(t, sig, 'k', label='true signal with trend')
ax1.set_title('10 Hz and 20 Hz sinusoids')

colors = ['b','green','orange']

for i,Wn in enumerate([0.1,1,10]):
    sos = signal.butter(10, Wn, 'hp', fs=1000, output='sos')
    filtered = signal.sosfilt(sos, sig)
    ax1.plot(t, filtered, label='filtered ('+str(Wn)+' Hz)', color=colors[i])
    
ax1.legend()
ax1.set_xlabel('Time [seconds]')
plt.tight_layout()
```
